```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
#-  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -   -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  - 

#[2-9] 다음은 타이타닉 데이터를 DT로 학습하고 테스트하는 프로그램을 작성하는 과정이다. 물음에 답하시오. 문제 당 하나의 셀(cell)에 답안을 적으시오.

#2.(데이터 로드) 이러닝캠퍼스에서 titanic_train.csv 파일을 읽어 DataFrame 자료형으로 변수 df에 저장하시오.(1점)
df = pd.read_csv("titanic_train.csv")

```

```python

#3.(데이터 분리) train_test_split 함수를 이용해 학습데이터와 테스트데이터를 나누시오. 단, test_size = 0.2, random_state = 1 로 설정하시오.(1점)
df_labels = df['Survived']
df.drop('Survived',axis=1,inplace=True)



x_train,x_test,y_train,y_test = train_test_split(df,df_labels, test_size = 0.2, random_state = 1)
x_train
```

```python


#4.(칼럼 삭제) 학습데이터에서 PassengerId, Name, Ticket, Cabin 칼럼을 없애시오.(1점)

x_train = x_train.drop(['PassengerId','Name','Ticket','Cabin'],axis =1)
print(x_train)
```

```python
#5.(결측치 처리) 학습데이터에서 Age의 결측치는 학습데이터 Age값의 평균으로 대체하고, Embarked 칼럼의 결측치는 문자 ‘N’으로 대체하시오.(1점)
x_train['Age']=x_train['Age'].fillna(x_train['Age'].mean())
x_train['Embarked'] = x_train['Embarked'].fillna('N') 
x_train.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
#6.(문자열 처리) 학습데이터에서 Sex와 Embarked 변수에 대해 레이블 인코딩을 수행하고, 레이블 인코더를 le_Sex, le_Embarked 변수에 각각 저장하시오.(1점)


encoder = LabelEncoder()
le_sex = encoder.fit(x_train['Sex'])
le_sex = encoder.transform(x_train['Sex'])
le_Embarked = encoder.fit(x_train['Embarked'])
le_Embarked = encoder.transform(x_train['Embarked'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
#7.(학습) 다음 지시에 따라, GridSearchCV를 이용하여 DecisionTreeClassifier를 학습하고 결과를 출력하시오.(5점)

# - DecisionTreeClassifier 생성 시, random_state 값은 1로 설정하시오.

# - 하이퍼 파라미터는 두 개만 튜닝(tuning)한다. max_depth는 3, 4, 5 중에서, min_sample_split이 2, 3, 4 중에서 최적의 하이퍼 파라미터를 찾으시오.

# - cross validation의 fold 수는 5로 설정하시오.

# - 최적 하이퍼 파라미터와 CV 최고 정확도를 출력하시오

# - 최적의 하이퍼 파라미터에 대해 분류기를 (다시) 학습하여 best_dt라는 변수에 저장하시오.

dtree = DecisionTreeClassifier(random_state=1)
parameters = {'max_depth':[3,4,5],'min_sample_split':[2,3,4]}
grid_tree = GridSearchCV(dtree, param_grid=parameters, scoring='accuracy',cv=5)

grid_tree.fit(x_train,y_train)


print("최적 하이퍼 파라미터",grid_tree.best_params_)
print("CV 최고 정확도:",grid_tree.best_score_)

best_dt = grid_tree.best_estimator_

```

```python

#8.(테스트 데이터 전처리) 테스트 데이터를 적절히 전처리 하시오.(5점)
x_test['Sex'] = le_sex
x_test['Embarked'] = le_Embarked


#9.(성능 측정) 테스트 데이터에 대한 best_dt 알고리즘 성능을 측정하여 출력(print)하시오.(1점)

accuracy = accuracy_score(y_test,best_dt)
print("테스트 데이터에 대한 성능:",accuracy)
```
