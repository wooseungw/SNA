```python
#(10) 나이가 20세 미만 남자들의 정보를 모두 출력하시오.

under_20 = df.loc[df['Age']<21]

under_20_mans = under_20[under_20['Sex'] == 'male']
under_20_mans
```

```python
#(11) df를 나이에 관해 오름차순 정렬하시오.
df.sort_values(by=['Age'],inplace=True)
```

```python
#(12) 산 사람과 죽은 사람의 나이와 요금의 평균을 구하시오.
survived_df = df[df['Survived'] == 1]
unsurvived_df = df[df['Survived'] == 0]
survived_df_age = survived_df['Age']
survived_df_fare = survived_df['Fare']
print("산사람의 평균 나이:",survived_df_age.mean()," 산 사람의 평균 요금:",survived_df_fare.mean())
unsurvived_df_age = unsurvived_df['Age']
unsurvived_df_fare = unsurvived_df['Fare']
print("산사람의 평균 나이:",unsurvived_df_age.mean()," 산 사람의 평균 요금:",unsurvived_df_fare.mean())
```

```python
#(13) 함수 isna를 이용하여 칼럼별 결측 데이터 개수를 출력하시오.

isna_df = df.isna()
isna_df.sum()
```

```python
#(14) Cabin의 결측치를 C00로 채우고, Age의 결측치는 median, Embarked는 S로 채우시오. *심화: Embarked를 채울 때 S 대신 mode함수를 이용해서 최빈값(mode, 제일 많이 등장한값)으로 채우는 코드를 작성하면 더 좋다.

df['Cabin'] = df['Cabin'].fillna('C00')
df['Age'] = df['Age'].fillna(df['Age'].median())
mode_embarked = df['Embarked'].mode()
df['Embarked'] = df['Embarked'].fillna(mode_embarked[0])
df
```

```python
#(15) payment 칼럼을 만들고 apply lambda 식을 사용하여 Fare를 10달러 미만 낸 사람은‘Cheap’, 10달러 이상 30달러 미만 낸 사람은 ‘Normal’, 30달러 이상낸 사람은 ‘Expensive’라고 적으시오.

df['payment'] = df['Fare'].apply(lambda x : 'Expensive'if x >= 30 else ('Normal' if x >= 10 else 'Cheap' ))
df['payment']
```
