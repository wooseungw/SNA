### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

* 앞에서 csv 파일을 읽어서 DataFrame으로 저장했다. 이번에는,
* 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환(저장)해 보자. 
* 거꾸로도 해 보자.

```python
# 리스트 -> DataFrame 생성: pd.DataFrame( )

col_name1=['c1'] # 컬럼 한 개 필요
list1 = [1, 2, 3]

df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)
```

```python
# ndarray -> DataFrame 생성

import numpy as np
array1 = np.array(list1) # 컬럼에 입력할 데이터(넘파이 배열)
print('array1 shape:', array1.shape )

df_array1 = pd.DataFrame(array1, columns=col_name1) # 동일한 컬럼명으로 지정
print('1차원 ndarray로 만든 DataFrame:\n', df_array1) # 위의 df_list1와 같다
```

```python
# 2행x3열 리스트 -> DataFrame  (열이 3개니까 컬럼이 3개 필요)

list2 = [[10, 20, 30],
         [11, 12, 13]]

# 3개의 칼럼명이 필요함.
col_name2=['c1', 'c2', 'c3']

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
```

```python
# 2행x3열 ndarray -> DataFrame

array2 = np.array(list2) # ndarray 데이터
print('array2 shape:', array2.shape )

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2) # df_list2와 같음
```

```python
# 딕셔너리 -> DataFrame

# Key는 컬럼명으로 매핑, Value는 리스트 형(또는 ndarray)
# 딕셔너리의 value들의 길이는 모두 같아야 한다

dict = {'c1':[10, 11], 'c2':[20, 12], 'c3':[30, 13]}

df_dict = pd.DataFrame(dict)                  # columns= 옵션을 사용하지 않는다
print('딕셔너리로 만든 DataFrame:\n', df_dict) # df_array2, df_list2와 같음
```

#### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 (거꾸로) 변환하기

```python
# DataFrame-> ndarray로 변환, values

array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
```

```python
# DataFrame-> 리스트로 변환, tolist()

list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)
```

```python
# DataFrame-> 딕셔너리로 변환, to_dict()

dict3 = df_dict.to_dict('list') # 괄호는 딕셔너리 value값 형태 지정을 위해 사용. list형태로.
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### DataFrame에서 row 삭제: drop( ... , axis=0)

```python
# print문으로 예쁘게 출력하기 위한 셋팅 > 일단 지나가자
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)

print('#### before axis 0 drop ####')
print(df.head())
```

```python
df.drop([0,1,2], axis=0, inplace=True) # inplace이면 df가 바뀐다

print('#### after axis 0 drop ####')
print(df.head())
```

```python
df2 = df.drop([3], axis=0) # inplace가 False이면, df 자신은 안 바뀐다, df2는 바뀜
print(df.head(), end='\n\n')
print(df2.head())
```

- axis: DataFrame의 로우 삭제(drop)시 axis=0으로, 칼럼 삭제시 axis=1로 설정
- 원본 DataFrame은 유지하고, 칼럼이나 삭제한 DataFrame을 새로 갖고 싶다면 inplace=False로 설정(디폴트 설정임)
- 원본 DataFrame을 유지하지 않고 드롭시키고 싶은 경우, inplace=True로 설정(제자리(in place)에, 이곳에, 적용해라)
- df = df.drop('Age_0', axis=1, inplace=False)는 df.drop('Age_0, axis=1, inplace=True)와 같다

### apply lambda 식으로 데이터 가공

```python
def get_square(a):
    return a**2

print('3의 제곱은:',get_square(3))
```

```python
lambda_square = lambda x : x ** 2
print('3의 제곱은:',lambda_square(3))
```

```python
a=[1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult' )
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 
                                                                                  'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성. 
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정. 
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
    
```
