```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('./data/시도별 전출입 인구수.xlsx',engine='openpyxl')
df.head(10)
```

```python
df=df.fillna(method='ffill')#ffill앞의정보로 채운다.
df.head()
print(df['전출지별'])
print(df['전출지별']=='서울특별시')#서울특별시와 비교한 결과
```
