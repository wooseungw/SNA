```python
#(16)

import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([2.4, 1.2, 3.5, 2.1, 4.7])  # 데이터


buff = 0

#정규분포 평균과 분산의 최대우도 추정치
mu = np.average(x_data)
sigma = np.var(x_data)
for i in x_data:
    buff += (i-mu)**2
mle = np.std(x_data)
print(mu,sigma,mle)

#우도와 대수우도
def pdf(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2)) 


print("우도",np.prod(pdf(x_data, mu, sigma)))

print("대수우도",np.sum(np.log(pdf(x_data, mu, sigma))))

#분포함수
y = np.zeros(5)


x_pdf = np.linspace(-1, 9)
y_pdf = pdf(x_pdf, mu, sigma)

plt.scatter(x_data, y)
plt.plot(x_pdf, y_pdf)
plt.show()
```

```python
import numpy as np
P=[0.35, 0.25, 0.1, 0.3]
Q=[0.32, 0.23, 0.11, 0.34]
def cross_entropy (P, Q):
    # 이 부분을 작성하세요. 
    buff = 0
    for i in range(len(P)):
        buff += P[i]*np.log(Q[i])
        
        
    return -buff

print("Cross Entropy = ", cross_entropy(P, Q))
```
