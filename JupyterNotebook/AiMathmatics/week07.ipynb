(1) 시그모이드 함수의 도함수의 특징을 설명하시오. x 의 범위에 따라 어떻게 달라지는가? 도함수 값의 최댓값을 구하시오.

시그모이드 함수를 f(x)라고 할 때 도함수는 (1-f(x))f(x)

도함수의 최대값이 1보다 작음 x가 0일때 최대값을 가지고 x가 -무한대,무한대로 갈때 점점 작아진다.

(2) 하나의 그림에 다음 두 함수를 그리시오.  의 범위를 0.01부터 5까지 설정하시오.

```python
#최소값을 구하는 법을 알면 최대값을 구할수있다.
#f(x)에 -만 붙히면 f(x)에서 최소값이 된 부분이 -f(x)에서 최대값이 된다.
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x)

def g(x):
    return np.log(x)



x = np.linspace(0.01,5)
y1 = f(x)
y2 = g(x)

plt.plot(x,y1)
plt.plot(x,y2)

plt.grid
plt.show() 
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3*x**4-8*x**3-6*x**2+24*x -1

def f_prime(x):
    return 12*x**3-24*x**2-12*x+24

eta = 0.001
x = -0.7

record_x = []
record_y = []

for i in range(30):
    y = f(x)
    record_x.append(x)
    record_y.append(y)
    x -= (eta * f_prime(x))
    
x_len = np.linspace(-2,2)
y_len = f(x_len)

plt.plot(x_len,y_len,linestyle="dashed")
plt.scatter(record_x,record_y,c="orange")
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)

print("Golbal minima:",np.round(x,4))
print("Golbal minimum:",np.round(y,4))

print("Local minima:",np.round(record_x[-1],4))
print("Local minimum:",np.round(record_y[-1],4))
plt.grid()
plt.show()
print(np.round(record_x,4))
print(np.round(record_y,4))
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -(3*x**4 - 8*x**3 - 6*x**2 + 24*x -1)

def f_prime(x):
    return -(12*x**3 - 24*x**2 - 12*x + 24)

eta = 0.001
x = -0.5

record_x = []
record_y = []

for i in range(20):
    y =f(x)
    record_x.append(x)
    record_y.append(y)
    x += eta * f_prime(x)
    
x_len = np.linspace(-2,2)
y_len = f(x_len)

plt.plot(x_len,y_len,linestyle="dashed")
plt.scatter(record_x,record_y,c="orange")
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)

print("Golbal maxima:",np.round(x,4))
print("Golbal maximum:",np.round(y,4))

print("Local maxima:",np.round(record_x[-1],4))
print("Local maximum:",np.round(record_y[-1],4))
plt.grid()
plt.show()
print(np.round(record_x,4))
print(np.round(record_y,4))
```
