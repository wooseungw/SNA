```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.arange(10,dtype="int32")

f = f.reshape(2,5)
print(np.shape(f))
print(f+3)
print(len(f))
g = np.ones(10,dtype="int32")
g = g.reshape(2,5)

#행렬의 연산
print(f+g,"\n")
#단순히 같은 요소끼리 곱한것
print(f*g,"\n")

f[:,0] = np.array([4,8])

print(f[1,:])
```

```python
def my_function(x):
    y = x * x+ 2+1
    return y

b = my_function(f)

print(b)

print(my_function(np.array([1,2,4,2,5,2,2])))
```

```python
print(f)
rsum = np.sum(f,axis=0)#row 방향의 합
csum = np.sum(f,axis=1)#collumn 방향의 합
print(rsum)
print(csum)
```

```python
A = np.array([
    [70, 80, 60],
    [50, 80, 90],
    [50, 60, 100],
    [50, 75, 90]])


sum = np.sum(A,axis=1)
avg = np.average(A,axis=1)
min = np.min(A,axis=0)
max = np.max(A,axis=0)



print("총합:",sum)
print("평균:",np.round(avg,0))
print("최고점수(국영수):",max)
print("최저점수(국영수):",min)
```
