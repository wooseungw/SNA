(4)다음 함수의 그래프와 x = 2 에서의 접선을 그리는 파이썬 프로그램을 작성하시오.
$$f(x) = -x^2 + x + 3$$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return -x**2+x+3

def func_dif(x):
    return -2*x+1
    

x = np.linspace(-5,5)
y = func(x)

x1 = 2

y_t = func_dif(x1)*x + func(x1) - func_dif(x1)*x1


plt.plot(x, y, label="y")
plt.plot(x, y_t, label="y_t")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
