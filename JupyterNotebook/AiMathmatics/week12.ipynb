```python
import numpy as np
import matplotlib.pyplot as plt 

a0_true = 2
a1_true = -1

X = np.linspace(-1, 2)
Y = a0_true + a1_true*X + np.random.randn(len(X)) # 노이즈 첨가
Y2 = a0_true + a1_true*X
plt.scatter(X, Y) # 우리에게 주어진 데이터
plt.plot(X,Y2)
plt.show()
# 데이터 값 확인
#for x, y in zip(X, Y):
# print(np.round(x-2, 2), np.round(y,2)
```

```python
def f(x, a0_, a1_): # 모델(함수) 식
    return a0_ + a1_*x
eta = 0.01 # 하이퍼 파라미터
epoch = 1000
# a0, a1 초기화(initialization)
a0, a1 = np.random.randn(2)
print(a0, a1)
E = np.zeros(epoch) # E를 그려 E가 정말 감소하는지 확인하기 위함
for n in range(epoch):    
# a0 업데이트
    a0 = a0 + eta*np.sum(f(X,a0,a1)-a0-a1*X)
# a1 업데이트
    a1 = a1 + eta*np.sum(f(X,a0,a1)-a0-a1*X)
# E값 저장    
    E[n] = (1/2)*np.sum(f(X,a0,a1)-(a0+a1*X))**2
# 값 확인. a0, a1의 참 값(true value)와 비슷한가? 잘 추정하는가?
print(np.round(a0, 4), np.round(a1, 4))
plt.plot(X, a0+a1*X) # 추정한 직선의 그래프

Y = a0_true + a1_true*X
plt.figure(1); plt.plot(X, Y, color='orange') #True 직선의 그래프
plt.figure(2)
plt.plot(range(20), E[:20])
plt.show()
```
