```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#1-1
x = np.linspace(-5,5)
y1  = np.sqrt(x) + 1

plt.title("Lecture 02")
plt.grid()

plt.plot(x,y1,label = "y = x^0.5 + 1")


plt.legend()
plt.show()
```

```python
#1-2
x = np.linspace(-5,5)
y2 = x**3- 2*x**2 -3*x + 4

plt.title("Lecture 02")
plt.grid()

plt.plot(x,y2,label = "y = x^3 - 2x^2 -3x + 4")

plt.legend()
plt.show()
```

```python
#1-3
x = np.linspace(-2*np.pi,2*np.pi)
y_cos = np.cos(x)
y_sin = np.sin(x) 

plt.title("Lecture 02")
plt.grid()

plt.plot(x,y_sin,label = "sin(x)")
plt.plot(x,y_cos,label = "cos(x)")

plt.legend()
plt.show()
```

```python
#1-4
x = np.linspace(-5,5)
y3 = np.abs(x**2 - 4)

plt.title("Lecture 02")
plt.grid()

plt.plot(x,y3,label = "|x^2+4|")

plt.legend()
plt.show()
```
