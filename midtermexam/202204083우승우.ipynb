```python
# %matplotlib lnline

import matplotlib.pyplot as plt
import numpy as np




def f(x):
    return (1/5)*(3 * x**4 - 4*x**3-12*x**2+12)

def f_prime(x):
    return (1/5)*(12*x**3 - 12*x**2 - 24*x)

#1. x=1에서 접선을 그리는 코드를 작성하고 그림을 출력하시오.(4점)
x = np.linspace(-5,5)
a = 1
y1 = f(x)
y2 = f_prime(a)*x-f_prime(a)*a+f(a)

plt.plot(x,y1,linestyle="dashed")
plt.plot(x,y2)



#2. 최급강하법을 이용하여, global minimum과 global minima를 구하는 코드를 작성하시오.(3.5점)

eta = 0.001
x1 = 2
record_x = []
record_y = []
for i in range(30):
    y = f(x1)
    record_x.append(x1)
    record_y.append(y1)
    x -= eta*f_prime(x1)
    
#plt.scatter(record_x,record_y)

print("local minima:",record_x[-1],"local minimum",record_y[-1])    

#3. 최급강하법을 이용하여, local minimum과 local minima를 구하는 코드를 작성하시오. local minima가 여러 개일 경우, 하나만 구하는 코드 작성. (3.5점)

x2 =4
record_local_x = []
record_local_y = []
for i in range(30):
    y = f(x2)
    record_local_x.append(x2)
    record_local_y.append(y1)
    x -= eta*f_prime(x2)
    
    
print("local minima:",record_local_x[-1],"local minimum",record_local_y[-1])
plt.show()
```
